```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sec_a = np.array([75, 65, 73, 68, 72, 76])
sec_b = np.array([90, 47, 43, 96, 93, 51])
no = np.array([1, 2, 3, 4, 5, 6])
```

```python
mean = np.mean(sec_b)
mean
```

```python
plt.figure(figsize=(10, 3))
```

```python
plt.scatter(sec_a, no, label="sec A")
plt.scatter(sec_b, no, color="red", label="sec B")
```

```python
plt.plot([70, 70, 70, 70, 70, 70], no, c="blue", label="Mean")
```

```python
mean_a = np.mean(sec_a)
mean_b = np.mean(sec_b)
```

```python
mean_dev_a = np.mean(np.abs(sec_a - mean_a))
mean_dev_b = np.mean(np.abs(sec_b - mean_b))
```

```python
std_dev_a = np.std(sec_a)
std_dev_b = np.std(sec_b)
```

```python
variance_a = np.var(sec_a)
variance_b = np.var(sec_b)
```

```python
range_a = np.max(sec_a) - np.min(sec_a)
range_b = np.max(sec_b) - np.min(sec_b)
```

```python
print("Section A Statistics:")
print(f"Mean: {mean_a}")
print(f"Mean Deviation: {mean_dev_a}")
print(f"Standard Deviation: {std_dev_a}")
print(f"Variance: {variance_a}")
print(f"Range: {range_a}")

print("\nSection B Statistics:")
print(f"Mean: {mean_b}")
print(f"Mean Deviation: {mean_dev_b}")
print(f"Standard Deviation: {std_dev_b}")
print(f"Variance: {variance_b}")
print(f"Range: {range_b}")
```
